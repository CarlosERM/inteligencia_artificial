```python
import random
TAMANHO_DA_POPULACAO = 100

GENES = '''abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOP
QRSTUVWXYZ 1234567890, .-;:_!"#%&/()=?@${[]}'''

''' Genes válidos. '''

ALVO = "Eu amo Ciencia da Computacao!"
'''A String alvo que deve ser gerada'''

class Individuo(object):
    '''Essa classe representa um indivíduo numa população.'''
    def __init__(self, cromossomo):
        self.cromossomo = cromossomo
        self.aptidao = self.calcular_aptidao()
    
    @classmethod
    def mutar_genes(self):
        '''Cria genes aleatórios para a mutação.'''
        global GENES
        gene = random.choice(GENES)
        return gene
    
    @classmethod
    def cria_cromossomo(self): 
        '''Cria um cromossomo ou uma cadeia de Genes.'''
        global ALVO
        tamanho = len(ALVO)
        return [self.mutar_genes() for _ in range(tamanho)]
    
    def reproduzir(self, parente2):
        '''Produz filho.'''

        cromossomo_do_filho = []

        for gp1, gp2 in zip(self.cromossomo, parente2.cromossomo):

            probabilidade = random.random()
            
            '''Probabilidade aleatória.'''
            if probabilidade < 0.45:
           
            # Se for menor que 0.45, insere gene do pai 1.
                cromossomo_do_filho.append(gp1)
            elif probabilidade < 0.90:
            # Se estiver entre 0.45 e 0.90, insere gene do pai 2.
                cromossomo_do_filho.append(gp2)
            else:
            # Insere um gene mutado.
            # Para manter a diversidade. 
                cromossomo_do_filho.append(self.mutar_genes())
        return Individuo(cromossomo_do_filho)

    def calcular_aptidao(self):
        '''
            Calcula a aptidão.
            Nesse aqui é o número de caracteres que se diferem do valor alvo.
        '''
        global ALVO
        aptidao = 0
        for cromossomo_letra, alvo_letra in zip (self.cromossomo, ALVO):
            if cromossomo_letra != alvo_letra: aptidao += 1
        return aptidao

def main():
    global TAMANHO_DA_POPULACAO

    # Geração atual.
    geracao = 1

    achou = False
    populacao = []

    # Cria uma população inicial.

    for _ in range(TAMANHO_DA_POPULACAO):
        cromossomo = Individuo.cria_cromossomo()
        populacao.append(Individuo(cromossomo))

    while not achou:

        # Ordena a população em ordem crescente de pontos de aptidão.
        populacao = sorted(populacao, key = lambda x:x.aptidao)

        # Se o indivíduo com a o menor valor de aptidão 
        # é encontrado, então nós alcançamos o alvo e quebramos
        # o loop.

        if populacao[0].aptidao <= 0:
            achou = True
            break

        nova_geracao = []

        # Vai fazer o Elitismo. 10% da população mais apta
        # vai para a próxima geração.

        s = int((10*TAMANHO_DA_POPULACAO)/100)
        nova_geracao.extend(populacao[:s])

        # 50% da população mais apta vai se reproduzir
        # e ter filhos.
        s = int((90*TAMANHO_DA_POPULACAO)/100)
        for _ in range(s):
            p1 = random.choice(populacao[:50])
            p2 = random.choice(populacao[:50])
            filho = p1.reproduzir(p2)
            nova_geracao.append(filho)
        
        populacao = nova_geracao

        print("Geração: {}\tString: {}\t Aptidão: {}".format(geracao, "".join(populacao[0].cromossomo), populacao[0].aptidao)
            )
        geracao += 1
    print("Geração: {}\tString: {}\t Aptidão: {}".format(geracao, "".join(populacao[0].cromossomo), populacao[0].aptidao)
    )
        
if(__name__ == '__main__'):
    main()
```
